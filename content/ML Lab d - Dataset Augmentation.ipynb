### 1. Rotate BCC cell and compute cell properties

Create a BCC metamaterial cell, and rotate it by 90, 180, and 270.

```python
# origianl without rotation.
pattern0 = generate_BCC_unitcell(a=30, r1=0.10, r2=0.15, r3=0.20, r4=0.25, show=False)
# clock-wise rotate by 90 degress.
pattern90 = generate_BCC_unitcell(a=30, r1=0.25, r2=0.10, r3=0.15, r4=0.20, show=False)
# clock-wise rotate by 180 degress.
pattern180 = generate_BCC_unitcell(a=30, r1=0.20, r2=0.25, r3=0.10, r4=0.15, show=False)
# clock-wise rotate by 270 degress.
pattern270 = generate_BCC_unitcell(a=30, r1=0.15, r2=0.20, r3=0.25, r4=0.10, show=False)

# plot cell patterns
imgs   = [pattern0, pattern90, pattern180, pattern270]; titles = ["0", "90", "180", "270"]
fig, axs = plt.subplots(1, 4, figsize=(8, 2))
for ax, im, t in zip(axs, imgs, titles):
    ax.imshow(1-im, cmap="gray"); ax.set_title(t); ax.axis("off")
plt.tight_layout(); plt.show()
```

Compute the effective properties of these four metamaterial cells using homogenisation method.

```python
# Unit Cell Geometry
lx = 1; ly = 1; phi = 90

# Material Property
E = 100; nu = 0.3
lambda_vals, mu_vals = compute_lambda_constants(E, nu)

# Compute the property features for each pattern
patterns = { "rot_0":   pattern0, "rot_90":  pattern90, "rot_180": pattern180, "rot_270": pattern270 }

# Run homogenisation and record cell properties
rows = []
for name, pattern in patterns.items():
    D = homogenise(lx, ly, lambda_vals, mu_vals, phi, pattern)
    vf = np.count_nonzero(pattern) / pattern.size
    feat = [D[0,0], D[0,1], D[0,2], D[1,1], D[1,2], D[2,2], vf] # Collect property vector: 6 stiffness components + volume fraction
    rows.append((name, feat))
```

List the metamaterial cells' properties for comparison.

```python
print_property_table(rows)
```

### 2. Data augmentation based on cell rotation

Load the original dataset.

```python
# --- Load the dataset as numpy array
data  = np.load("dataset_2D_BCC_500.npz")
prop  = data["prop"]    # input features
param = data["param"]   # target outputs

# --- Determine dimensions
N, in_dim  = prop.shape
_, out_dim = param.shape
```

We augment the dataset by rotating the structures and updating both the node parameters and their target properties accordingly.

```python
P, Y = param, prop

# --- 0 rotation
P0, Y0 = P, Y

# --- 90 rotation
P90 = P[:, [3,0,1,2]]
Y90 = np.c_[ Y[:,3], Y[:,1], -Y[:,4], Y[:,0], -Y[:,2], Y[:,5], Y[:,6] ]

# --- 180 rotation
P180 = P[:, [2,3,0,1]]
Y180 = Y

# --- 270 rotation
P270 = P[:, [1,2,3,0]]
Y270 = np.c_[ Y[:,3], Y[:,1], -Y[:,4], Y[:,0], -Y[:,2], Y[:,5], Y[:,6] ]

# --- Stack all four versions
param_aug = np.r_[P0, P90, P180, P270]
prop_aug  = np.r_[Y0, Y90, Y180, Y270]

print(f"Original samples : {param.shape[0]}")
print(f"Augmented samples: {param_aug.shape[0]}")
```

### <span style="color:lightblue;">MLP training and testing  </span>

```python
# Set augemented dataset as the dataset to train, valid and test the model
param = param_aug
prop = prop_aug
```

```python
# --- Split dataset
prop_train, prop_val, prop_test, param_train, param_val, param_test = split_prop_param(
    prop, param, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15, print_summary=False)
```

```python
# --- Scale dataset
prop_train, prop_val, prop_test, param_train, param_val, param_test, in_scaler, out_scaler = scale_datasets(
    prop_train, prop_val, prop_test, param_train, param_val, param_test,
    scale_inputs=True, scale_outputs=True, feature_range=(-1, 1), print_summary=False)
```

```python
# --- Convert to Torch tensors
X_train, Y_train, X_val, Y_val, X_test, Y_test = map(
    lambda a: torch.from_numpy(a).float(),
    (prop_train, param_train, prop_val, param_val, prop_test, param_test))
```

```python
# --- Data loader
BATCH_SIZE = 256 # number of samples per mini-batch
train_loader = DataLoader(TensorDataset(X_train, Y_train), batch_size=BATCH_SIZE, shuffle=True)
val_loader   = DataLoader(TensorDataset(X_val,   Y_val),   batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(TensorDataset(X_test,  Y_test),  batch_size=BATCH_SIZE, shuffle=False)
```

```python
# --- Defome model archietecture
HIDDEN = (8, )         # e.g., (), (64,), (128,64), (256,128,64), ...
DROPOUT = 0.0
model = MLP(in_dim=in_dim, hidden=HIDDEN, out_dim=out_dim, dropout=DROPOUT)
```

```python
# --- Train model
model, train_losses, val_losses = train_model(
    model,
    train_loader,
    val_loader,
    epochs=300, # training epoch
    patience=20, # early stopping if no validation improvement
    criterion=nn.MSELoss(), # Mean Squared Error (MSE) loss for regression
    optimizer=torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4), # Adam optimizer  efficient and commonly used
)
```

```python
# --- Evaluate model
y_true_plot, y_pred_plot, mae, rmse, r2 = evaluate_model(
    model, test_loader, out_dim=out_dim, out_scaler=out_scaler, scale_outputs=True, print_summary=False,)
plot_parity_plots(
    y_true_plot=y_true_plot, y_pred_plot=y_pred_plot, mae=mae, r2=r2, out_dim=out_dim,)
```
